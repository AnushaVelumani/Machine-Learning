#### Generate a training data set of size 100 similar to that used in Exercise 1 .4. Generate a test data set of size 10, 000 from the same process. To get g, run the algorithm above with n = 100 on the training data set, until a maximum of 1, 000 updates has been reached . Plot the training data set, the target function f, and the final hypothesis g on the same figure. Report the error on the test set.

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['axes.unicode_minus'] = False 

#generate a data set of 100. 
#for simplicity, 50 in the first quadrant, another 50 in the third quadrant 

def generate_points(N):
    X1 = []
    Y1 = []
    X2 = []
    Y2 = []

    for i in range(N):
        X1.append(random.uniform(0,1))
        Y1.append(random.uniform(0,1))
        X2.append(random.uniform(-1,0))
        Y2.append(random.uniform(-1,0))
        
    #label the data
    data1 = [np.array([1,X1[i],Y1[i],1]) for i in range(50)]
    data2 = [np.array([1,X2[i],Y2[i],-1]) for i in range(50)]
    data = data1 + data2
    
    return data
 
    

X = []

Y = []

for i in range(100):
    Y.append(data[i][3])
#print(Y)

for i in range(100):
    X.append(data[i][:3])
    
    
#The perceptron algorithm to find the weight and calculate exact target function
def Perceptron(X, Y):

    weights = [0,0,0]
    n = 100
    epoch = 10
    iteration = 0

    for k in range(epoch):
        for i, n in enumerate(X):
            predicted_y = np.sign(np.dot(np.transpose(weights), X[i]))    # y' = w^T *X  
            actual_y = Y[i]
            if (predicted_y * actual_y) >1:
                weights = weights
            else:
                iteration = iteration + 1
                weights = weights + n*(actual_y-predicted_y)*X[i]
                   
    print("No.of iterations= " + str(iteration) )
    return weights


z = Perceptron(X,Y)
print(z)

w0 = z[0]
w1 = z[1]
w2 = z[2]

slope = -w1/w2
intercept = -w0/w2

generate_points(N=100)

plt.figure(figsize=(8,6), dpi=60)

def abline(slope, intercept):
     # Plot a line from slope and intercept
     ax = plt.gca()
     x_value = np.array(ax.get_xlim())
     y_value = intercept + slope * x_value
     plt.plot(x_value, y_value, '--')

abline(slope,intercept)  

def plot_points(data):

    for i in range(100):
        if (data[i][3] == 1):
            plt.plot(data[i][1], data[i][2], 'ro')
        elif (data[i][3] ==-1):
            plt.plot (data[i][1], data[i][2], 'bo')
        
    plt.xlabel('X')
    plt.ylabel('Y') 
        
    plt.plot([-1,1], [1,-1], 'r-')
    
    
plot_points(data)

plt.show()
```
